## GitHub Actions Workflow

A GitHub Actions workflow was configured to automate Continuous Integration (CI). Whenever code is pushed to the repository or a pull request is created, the workflow automatically installs project dependencies, performs code quality checks using Flake8, and executes unit tests using Pytest. This ensures that the project remains stable and reduces the risk of introducing errors into the codebase.
